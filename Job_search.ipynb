## Updating JobCode.json (JobTest.json is json version of XML file)

```python
import os
import re
import pandas as pd
path = "C:/Users/Civic Homes/Desktop/Jobs/"
date_dict = {}
for file in os.listdir(path):
    if re.findall("AllJobs", file):
        date = re.findall("[0-9\-]+", file)[0]
        
        date = pd.to_datetime(date, format = "%d-%m-%Y")
        date_dict[date] = file
        
job_xml_path = date_dict[sorted(date_dict)[-1]]

import json
import xmltodict

with open(job_xml_path) as xml_file:
    data_dict = xmltodict.parse(xml_file.read())
    
out_file = open("C:/Users/Civic Homes/Desktop/Jobs/JobTest.json", "w")
json_data = json.dump(data_dict, out_file, indent = 6)
out_file.close()

data = json.load(open("C:/Users/Civic Homes/Desktop/Jobs/JobTest.json"))

try:
    jobs_list = data["Databuild"]["Jobs"][0]["JobsRecord"]
    contacts_list = data["Databuild"]["Contacts"][0]["ContactsRecord"]
except:
    jobs_list = data["Databuild"]["Jobs"]["JobsRecord"]
    contacts_list = data["Databuild"]["Contacts"]["ContactsRecord"]

job_code_list, address_list, customer_list = {}, {}, {}
for num, job in enumerate(jobs_list):
    res = jobs_list[num]|contacts_list[num]
    try:
        customer_list[res["Name"]] = [res["Name"], res["Site_Address"], res["Code"]]
    except KeyError:
        print(res)
        continue
        
    try:
        address_list[res["Site_Address"]] = [res["Name"], res["Site_Address"], res["Code"]]
    except KeyError:
        print(res)
        continue
        
    try:
        job_code_list[res["Code"]] = [res["Name"], res["Site_Address"], res["Code"]]
    except KeyError:
        print(res)
        continue
    
final_json = {"Customer":customer_list, "Address":address_list, "Job Code":job_code_list}
        
    
out_file = open("C:/Users/Civic Homes/Desktop/Jobs/JobCode.json", "w")
json_data = json.dump(final_json, out_file, indent = 6)
out_file.close()

data = json.load(open("C:/Users/Civic Homes/Desktop/Jobs/JobCode.json"))
```
