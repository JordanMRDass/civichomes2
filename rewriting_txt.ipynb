```python
import os
import pandas as pd
import re

path = "G:/My Drive/Estimator/Journal_Entry/"

instructions = '1. Take files from either "jnl_entries" or "jnl_entries_detailed"\n2. Correct the job codes for unknown variables (Excel spreadsheet will automatically calculate new sums and GL account codes)\n3. Place corrected excel spreadsheets in the "checked_corrected_jnl"\n'
space_detailed = "\n\nList of unknowns yet to be filled (detailed)\n\n"
big_line = "_"* 135
space = "\n\nList of unknowns yet to be filled\n\n"

path_of_excel = "G:/My Drive/Estimator/Journal_Entry/jnl_entries/"
path_of_excel_detailed = "G:/My Drive/Estimator/Journal_Entry/jnl_entries_detailed/"

def loop_through(dirmain):
    return [dir for dir in os.listdir(dirmain)]
        

def finder(path, dirtarget, dirname_actual):
    if dirname_actual == dirtarget:
        return path
    
    # If not in main
    initial_list = loop_through(path)
    for dirname in initial_list:
        try:
            result = finder(path + "/" + dirname, dirtarget, dirname)
            if result is not None:
                return result
        except NotADirectoryError:
            pass

def cleaning_excel(df):
    # Dropping null
    df_drop = df.dropna(subset = "Job Code")
    
    # Dropping cols
    df_drop_col = df_drop[['Job Code','Reference','Empty','Cost Center','Empty2','Credit','Debit','Date','Date2','Date3','GL_account','False','Comments','Empty3']].set_index("Job Code")
    
    return df_drop_col

def list_unknowns(detailed):
    if detailed == True:
        path = path_of_excel_detailed
        
    elif detailed == False:
        path = path_of_excel
        
    str_to_show = ""
    file_list = os.listdir(path)
    problem_str = "\nThese Addresses are not in DataBuild"
    for file in file_list:
        if re.findall('\~|\.ini', file) or file == 'unknowns' or file == 'knowns':
            continue
            
        df = pd.read_excel(path + file)
        
        df_clean = cleaning_excel(df)
        
        df_clean = df_clean.reset_index()
        
        unknowns_df = df_clean[df_clean["Cost Center"] == "unknown"]
        
        job_code_list = list(df_clean["Job Code"])
        
        num_exist = 0
        for job_code in job_code_list:
            if re.findall("\.xlsx", job_code):
                num_exist += 1
                address = re.sub("\.xlsx", "", job_code)
                problem_str += f"\n{num_exist}. {address}"
            
                
            
        if len(unknowns_df) == 0:
            continue
        str_to_show += f"\n{file} unknowns:"
        num = 1
        addition_of_job_unknowns = ""
        for job_code, cost_center, debit, GL_account, comments in zip(unknowns_df["Job Code"],unknowns_df["Cost Center"],unknowns_df["Debit"],unknowns_df["GL_account"],unknowns_df["Comments"]):
            addition_of_job_unknowns += f"\t{num}. Comments: {comments}\n\t   Debit: {debit}\n"
            num += 1
        
        str_to_show += "\n\t" + job_code + ":\n" + addition_of_job_unknowns
        str_to_show += "\n\n"
        
    problem_str += "\n"
    return str_to_show, problem_str

str_to_show_detailed, problem_str_detailed = list_unknowns(True)
str_to_show, problem_str = list_unknowns(False)

for file in os.listdir(path):
    if re.findall('\.txt', file):
        txt_file = file
        
print(path+txt_file)
f = open(path + txt_file,'w')
f.write(instructions + big_line + problem_str + big_line + space_detailed + str_to_show_detailed + big_line + space + str_to_show)
f.close()
print(instructions + big_line + problem_str + big_line + space_detailed + str_to_show_detailed + big_line + space + str_to_show)
```
